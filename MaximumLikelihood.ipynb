### 3. After two tossing

### 4. After three tossing
Assume I have tossed the coin for 3 times, and get the H, H, T

likelihood = p*p*(1-p)

```
p_1 = [i/11 for i in range(11)]
y = [p**2 * (1-p) for p in p_1]

plt.figure(figsize = (12, 6))
plt.plot(p_values, y)

max_value = max(y)
max_idx = np.argmax(y)
plt.plot(p_1[max_idx], [max_value], "ro")

plt.title("The posterior distribution after 3 tossing, H, H, T")
plt.show()
```

### 5. After 8 tossing
Assume I have tossed the coin for 8 times, and get 4 heads,

likelihood = p**4*(1-p)**4

```
p_1 = [i/11 for i in range(11)]
y = [p**4 * (1-p)**4 for p in p_1]

plt.figure(figsize = (12, 6))
plt.plot(p_values, y)
max_value = max(y)
max_idx = np.argmax(y)
plt.plot(p_1[max_idx], [max_value], "ro")
plt.title("The posterior distribution after 8 tossing, get 4 Heads")
plt.show()
```
