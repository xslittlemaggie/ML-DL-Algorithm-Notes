```
x = np.arange(0., 100., 1)

y = [ i** 2 - 1000 for i in x]
plt.figure(figsize=(10,10))
plt.plot(x, y)
plt.plot(60, 2600, "o")
#plt.plot(460/12., 0, 'o')
#plt.plot(x, (460 * x )/6. - (460 ** 2)/72 + 469, color = 'blue', label='2_itera')
#plt.plot(460/12., 469, 'o')
#plt.plot(x, 120 * x - 4600, color = 'red', label='1_itera')
plt.plot(32.2159420289855, 37.86692081495448, 'o')
plt.plot(x, 64.431884057971 * x - 2113.6007624448634, color = 'blue', label='3_itera')
plt.axhline(y = 0, color = 'black')
plt.ylim(-10000, 10000)
plt.legend()
plt.show()
```

```
def y(x):
  return 4 * (x + 1) ** 3
def b(x):
  return 12 * (x + 1) ** 2

def solve(x0):
  xi = x0
  for i in range(100):
    xi = - y(xi)/b(xi) + xi  
  return xi
```

```
solve(2.)
```
